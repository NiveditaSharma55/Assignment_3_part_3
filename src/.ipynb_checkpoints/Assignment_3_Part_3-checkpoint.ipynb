```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("train.csv")
df
```

```python
df.head()
```

```python
# Check data types
print(df.dtypes)
```

```python
# Summary statistics
print(df.describe())
```

```python
# To analyze the distribution of the variables YearBuilt and SalePrice 
#YearBuilt: Original construction date
#SalePrice - the property's sale price in dollars.
print(df[['YearBuilt', 'SalePrice']].describe())
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# Histogram for year Built
plt.hist(df['YearBuilt'], bins=20)
plt.xlabel('YearBuilt')
plt.show()
```

```python
# Histogram for Sales Price
plt.hist(df['SalePrice'], bins=20)
plt.xlabel('SalePrice')
plt.show()
```

```python
# Scatter plot for YearBuilt and SalePrice
plt.scatter(df['YearBuilt'], df['SalePrice'])
plt.xlabel('YearBuilt')
plt.ylabel('SalePrice')
plt.show()
```

```python
# Correlation coefficient for YearBuilt nad SalePrice
corr = df[['YearBuilt', 'SalePrice']].corr()
print(corr)
```

```python
#Visualize correlation To visualize the correlation between the variables, we can use a heatmap.
# Heatmap for YearBuilt nad SalePrice
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
#To analyze the distribution of the variables OverallCond and SalePrice 
#GarageArea: Size of garage in square feet
#SalePrice - the property's sale price in dollars.
print(df[['GarageArea', 'SalePrice']].describe())
```

```python
# Histogram for Garage Area
plt.hist(df['GarageArea'], bins=20)
plt.xlabel('GarageArea')
plt.show()
```

```python
# Creating a scatter plot between GarageArea and SalePrice 
plt.scatter(df['GarageArea'], df['SalePrice'])
plt.xlabel('Garage Area')
plt.ylabel('Sale Price')
plt.title('Scatter Plot of Garage Area vs. Sale Price')
plt.show()
```

```python
# Correlation coefficient for GarageArea and SalePrice
corr1 = df[['GarageArea', 'SalePrice']].corr()
print(corr1)
```

```python
#Visualize correlation To visualize the correlation between the variables, we can use a heatmap.
# Heatmap for GarageArea and SalePrice
sns.heatmap(corr1, annot=True, cmap='coolwarm')
plt.show()
```
